### 1. Investigate the domain architectures for the O category (largest expansion for Hexameta compared to the others)

### 1.1 pre-processing

Merge the interproscan file from the different species (hin, wb and spiro) in COG category O

```python
# pre-processing 
#display the raw interproscan file, merge all the O cog category files and crreate a file with protein id and specie column

import pandas as pd

interproscan_header = ['protein_id','Seq MD5 digest','Sequence length','Analysis', 'Signature accession', 
                           'Signature description','Start location', 'Stop location', 'Score', 'Status', 
                           'Date of run', 'Interpro annotations - accession', 'Interpro annotation - description', 
                           'optional: GO terms', 'optional: Pathway annotation']


df_hin_ss_group_O = pd.read_csv('../data/interproscan/HIN_ss_group_O.tsv', sep='\t', header=None, names = interproscan_header)
df_hin_ss_group_O_pids = df_hin_ss_group_O[['protein_id']].drop_duplicates()
df_hin_ss_group_O_pids['specie'] = 'hin'

df_spiro_ss_group_O = pd.read_csv('../data/interproscan/spiro_ss_group_O.tsv', sep='\t', header=None, names = interproscan_header)
df_spiro_ss_group_O_pids = df_spiro_ss_group_O[['protein_id']].drop_duplicates()
df_spiro_ss_group_O_pids['specie'] = 'spiro'

df_wb_ss_group_O = pd.read_csv('../data/interproscan/wb_ss_group_O.tsv', sep='\t', header=None, names = interproscan_header)
df_wb_ss_group_O_pids = df_wb_ss_group_O[['protein_id']].drop_duplicates()
df_wb_ss_group_O_pids['specie'] = 'wb'

df_O_pid_species_all = pd.concat([df_hin_ss_group_O_pids, df_spiro_ss_group_O_pids, df_wb_ss_group_O_pids], ignore_index=True)
df_O_pid_species_all = df_O_pid_species_all[df_O_pid_species_all['protein_id']!='0']
df_O_pid_species_all.to_csv('../results/pid_specie_cog_cat_O.tsv', sep='\t', index=False)

df_O_species_all = pd.concat([df_hin_ss_group_O, df_spiro_ss_group_O, df_wb_ss_group_O], ignore_index=True)
df_O_species_all.to_csv('../results/specie_cog_cat_O_interproscan.tsv', sep='\t', index=False)

```
